# Анализ столкновений птиц с воздушными судами



**Описание:**
   
Изучение влияния погоды на столкновения птиц с самолетами. Проанализировать данные для проведения дальнейшего исследования и выявить закономерности в данных. На основе исследования будет построена модель предсказания столкновений птиц с самолётами.
Стек технологий DA: Обработка сырых данных, Python (pandas, seaborn, plotly и другие библиотеки на ваш выбор). 
    
**Цель проекта:**
    
    Проанализировать данные + передать данные для дальнейших исследований ваших коллег из DS-направления 
Что нужно передать команде DS:
– проанализировать какие факторы коррелируют со столкновениями
- погода?
- аэропорт, трафик в аэропорте?
- на вылете или прилете?
- есть ли тренды?
– как можно агрегировать по году-месяцу эти данные чтобы создать признаки для прогнозирования количества столкновений?
На выходе нужны данные в виде таблицы csv и описание инсайтов
год, месяц
количество полетов
аэропорт, вылет, прилет
найденные признаки по колонкам

**Этапы и задачи, ТЗ проекта:**
    
1.	Предобработать данные, определить типы данных в таблице. 
2.	Обработать данные о погоде (WND, CIG, VIS, TMP, DEW, SLP), преобразовать их в отдельные столбцы.

Создадим дополнительную колонку с временем суток с такой разбивкой.Потом помотрим на разницу
с 0 до 6 часов — ночь
с 6 до 12 часов — утро
с 12 до 18 часов — день
с 18 до 24 часов — вечер

```python
# Создайте функцию для определения времени суток
def day_time(x):
    if pd.isna(x):
        return 'неизвестно'
    elif 0 <= x.hour < 6:
        return 'ночь'
    elif 6 <= x.hour < 12:
        return 'утро'
    elif 12 <= x.hour < 18:
        return 'день'
    else:
        return 'вечер'

# Примените функцию к столбцу 'time' и создайте новый столбец 'время_суток'
df_wn['day_time'] = df_wn['time'].apply(day_time)
```

```python
def categorize_species(species_id):
    if species_id.startswith(('Q', 'R', 'S', 'C')):
        return 'Ночные птицы'
    elif species_id == 'UNKB':
        return 'Неизвестные птицы'
    else:
        return 'Дневные птицы'

# Применить функцию к столбцу 'species_id' и создать новый столбец 'Category'
df_wn['сat_bird'] = df_wn['species_id'].apply(categorize_species)
```
