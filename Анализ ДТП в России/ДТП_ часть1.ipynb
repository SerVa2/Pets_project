```python
# Заплатка для Windows и новой библиотеки.

def read_geojson(filename):
    with open(f'{filename}', encoding= 'utf-8') as response:
        geo_json = json.load(response)
    return geo_json
```

```python
# Создадим функцию для исследования данных
def inform(data):
    display(data.head())
    print('Колонки')
    print(data.columns)
    
    print(50*'-')
    print('Содержимое колонок')
    print(20*'*')
    
    for col in data:
        print(col)
# Не все колонки могут быть прочитаны
# unique и nunique
        try:
            print(data[col].unique())
            print(data[col].nunique())
            
            
        except TypeError:
            print('Списковое исключение')
            
            print(data[col][:5])
            
           
    
        print(20*'*')
    print('Общая информация')
    print(data.info())
        
    print(50*"-")
    print('Пропуски')
    print(data.isna().sum())
    print(50*"-")
```

```python
def swap_coordinates(coord_str):
    if isinstance(coord_str, list):
        return [coord_str[1], coord_str[0]]
    
    match = re.match(r'\[(.*?), (.*?)\]', coord_str)
    if match:
        try:
            if match.group(1) and match.group(2):
                return [float(match.group(2)), float(match.group(1))]
        except ValueError:
            return None
    return None

# Применяем функцию к столбцу coordinates
df_base1['coordinates'] = df_base1['coordinates'].apply(lambda x: swap_coordinates(x) if x is not None else None)
```
