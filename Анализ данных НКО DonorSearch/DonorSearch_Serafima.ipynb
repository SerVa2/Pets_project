### **Cвязь между оценками доноров и количеством донаций.**

Для этого введем условную среднюю оценку донора - распределим средние оценки по диапазонам:
* **1:** интервал 0.0 - <1.5
* **2:** интервал 1.5 - <2.5
* **3:** интервал 2.5 - <3.5
* **4:** интервал 3.5 - <4.5
* **5:** интервал 4.5 - 5.0

```python
feedback_pivot_df = feedback_df.pivot_table(index='id_пользователя', values=['место_сдачи', 'оценка'], aggfunc={'оценка':'mean','место_сдачи':'count'})
feedback_pivot_df = feedback_pivot_df.rename(columns={'место_сдачи':'qty', 'оценка':'avg_mark'})

# Обобщим оценки на основе интервалов:
cut_bins = [0,1.5,2.5,3.5,4.5,6]
cut_labels = [1,2,3,4,5]
feedback_pivot_df['avg_mark'] = pd.cut(feedback_pivot_df['avg_mark'], bins=cut_bins, labels=cut_labels)

feedback_pivot_df.sort_values(by='qty', ascending=False)
```

```python
feedback_pivot_df['qty'].hist(bins=30)
plt.title('Распределение доноров по количеству донаций');
```

Подавляющее число доноров сдают кровь/плазму только один раз.

```python
feedback_pivot_df['avg_mark'].hist(bins=5)
plt.title('Распределение доноров\nпо усредненным оценкам обратоной связи');
```

```python
feedback_pivot_df.query('qty >= 10')['avg_mark'].hist(bins=5)
plt.title('Распределение оценок обратоной связи\nпо часто сдающим донорам');
```

Основная часть оценок - это усредненная 5. Что логично - если не нравится, человек вряд ли будет ходить часто сдавать кровь.
